```python
import numpy as np
import matplotlib.pyplot as plt

X = np.dot(np.random.rand(2, 2), np.random.rand(2, 200)).T
X_centered = X - np.mean(X, axis=0)

cov = np.cov(X_centered.T)
values, vectors = np.linalg.eig(cov)

plt.scatter(X_centered[:, 0], X_centered[:, 1], alpha=0.3)

for i in range(2):

    projection = np.dot(X_centered, vectors[:, i])
    p_min, p_max = projection.min(), projection.max()
    
    line = np.array([vectors[:, i] * p_min, vectors[:, i] * p_max])
    
    plt.plot(line[:, 0], line[:, 1], lw=3, label=f'PC{i+1}')

plt.axis('equal')
plt.legend()
plt.show()
```
