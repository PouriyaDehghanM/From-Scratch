```python
import numpy as np
import matplotlib.pyplot as plt

X = np.random.randn(100, 2)
y = (X[:, 0] + X[:, 1] > 0).astype(int)

w, b = np.zeros(2), 0
lr = 0.1
losses = []

for _ in range(1000):
    z = np.dot(X, w) + b
    y_pred = 1 / (1 + np.exp(-z)) 

    #Binary Cross-Entropy Loss
    loss = -np.mean(y * np.log(y_pred + 1e-9) + (1 - y) * np.log(1 - y_pred + 1e-9))
    losses.append(loss)
    
    
    dw = np.dot(X.T, (y_pred - y)) / len(y)
    db = np.sum(y_pred - y) / len(y)
    w -= lr * dw
    b -= lr * db


plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', alpha=0.6)
x_axis = np.array([X[:, 0].min(), X[:, 0].max()])
y_axis = -(w[0] * x_axis + b) / w[1] #  : w1x1 + w2x2 + b = 0
plt.plot(x_axis, y_axis, color='black', lw=2)
plt.title("Logistic Regression from Scratch")
plt.show()

x_losses = range(1000)
plt.plot(losses, x_losses)
```
