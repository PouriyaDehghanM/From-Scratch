```python
import numpy as np

X = np.linspace(1, 10, 20)

true_w = 2
true_b = 1

noise_level = 1.0 
noise = np.random.randn(len(X)) * noise_level

y = true_w * X + true_b + noise

w = 0.0
b = 0.0

lr = 0.01
epochs = 10000

n = len(X)

for i in range(epochs):
    y_pred = w * X + b
    error = y_pred - y

    dw = (2/n) * np.sum(error * X)
    db = (2/n) * np.sum(error)

    w -= lr * dw
    b -= lr * db

    if i % 500 == 0:
        mse = np.mean(error**2)
        print(f"Epoch {i}  MSE={mse:.2f}")

print("\nمدل نهایی:")
print(f"y = {w:.2f}x + {b:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y, color='blue', label='Data')

y_line = w * X + b
plt.plot(X, y_line, color='red', label='Model')

plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression From Scratch")
plt.legend()
plt.show()
```
