```python
import numpy as np

X_train = np.array([[1, 2], [2, 3], [3, 1], [6, 5], [7, 7], [8, 6]])
y_train = np.array([0, 0, 0, 1, 1, 1]) 

def knn_predict(new_point, k=3):

    distances = np.linalg.norm(X_train - new_point, axis=1)
    
    k_indices = np.argsort(distances)[:k]
    
    return np.bincount(y_train[k_indices]).argmax()

test_point = [1, 1]
print(f"predict: {knn_predict(test_point)}")
```

```python
import matplotlib.pyplot as plt

test_point = [5, 5]

plt.scatter(X_train[:,0], X_train[:,1], c=y_train)

plt.scatter(test_point[0], test_point[1], marker='*', s=200)

plt.title("KNN simple plot")
plt.show()
```
