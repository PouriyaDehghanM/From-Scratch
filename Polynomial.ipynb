```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 5, 10, 17, 26], dtype=float) 

degree = 2
lr = 0.001
epochs = 10000


phi = np.column_stack([X**i for i in range(degree+1)])

w = np.zeros(degree + 1)

n = len(X)

for _ in range(epochs):
    y_pred = phi @ w
    grad = 2/n * phi.T @ (y_pred - y)
    w -= lr * grad

print (w)

plt.scatter(X, y)
plt.plot(X, y_pred)
```
